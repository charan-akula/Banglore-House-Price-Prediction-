```python
# fro some of the strings we have - to represnt limits so we should remove them and take their mean
def remove_hypen(x):
    if "-" in x:
        lt=x.split("-")
        v1=float(lt[0])
        v2=float(lt[1])
        return ((v1+v2)/2)
    elif not(x.isnumeric()):
        return np.NAN
    elif x.isnumeric():
        return float(x)
       
        
        
        
```
